# Initial setup

1. Create unity catalog with different schemas following the medalion architecture. 
2. Ingest the data in to unity catalog
3. Create subfolders in volume for each type of data
4. Move data files to correct volume folder

```python
%sql
CREATE CATALOG IF NOT EXISTS marathos;

CREATE SCHEMA IF NOT EXISTS marathos.bronze;
CREATE SCHEMA IF NOT EXISTS marathos.silver;
CREATE SCHEMA IF NOT EXISTS marathos.gold;
CREATE SCHEMA IF NOT EXISTS marathos.information_schema;

CREATE VOLUME IF NOT EXISTS marathos.default.raw
```

```python
dbutils.fs.mkdirs("/Volumes/marathos/default/raw/historic_data") 
dbutils.fs.mkdirs("/Volumes/marathos/default/raw/new_data") 
dbutils.fs.mkdirs("/Volumes/marathos/default/raw/reference_data")
```

```python
dbutils.fs.mv(
    "/Volumes/marathos/default/raw/TWO_CENTURIES_OF_UM_RACES.csv",  
    "/Volumes/marathos/default/raw/historic_data/TWO_CENTURIES_OF_UM_RACES.csv"  
)

dbutils.fs.mv(
    "/Volumes/marathos/default/raw/countries.csv",
    "/Volumes/marathos/default/raw/reference_data/countries.csv"
)

dbutils.fs.mv(
    "/Volumes/marathos/default/raw/new_race_data.csv",
    "/Volumes/marathos/default/raw/new_data/new_race_data.csv"
)
```
